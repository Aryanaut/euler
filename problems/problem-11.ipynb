```python
# intM
for t in m:
    l = []
    for i in t:
        l = i.split()
    intM.append(l)
```

```python
# horizontal searching
hor = []
combinationsH = []
for x in intM:
    c = []
    for i in range(len(x)-4):
        c.append(x[i:i+4])
    combinationsH.append(c)
        
for el in combinationsH:
    for e in el:
        productH = 1 
        for l in e:
            productH = productH * int(l)
        hor.append(productH)
        
print(hor)
```

```python
# vertical searching
ver = []
combinationsV = []
c = []
for y in range(20):
    n = 0
    f = []
    for i in range(20):
        f.append(intM[i][y])
    c.append(f)
    
for el in c:
    for e in range(len(el)-4):
        combinationsV.append(el[e:e+4])

for combination in combinationsV:
    productV = 1
    for item in combination:
        productV = productV * int(item)
    ver.append(productV)
    
print(ver)
```
