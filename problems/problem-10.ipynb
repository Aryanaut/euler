## Problem 10
The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.

```python
import math
```

```python
def isPrime(n):
    if n > 1:
        for i in range(2, int(math.sqrt(n))+1):
            if (n % i) == 0:
                prime = False
                break
        else:
            prime = True
    else:
        prime = False
    return prime
```

```python
def getPrimes(end):
    primes = []
    for x in range(2, end):
        p = isPrime(x)
        if p == True:
            primes.append(x)
    return primes
```

```python
primes = getPrimes(2000000)
print(primes)
print(sum(primes))
```
